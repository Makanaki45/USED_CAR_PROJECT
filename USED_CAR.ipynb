```python
from matplotlib import pyplot as pt
```

```python
ax = car.Registration_Year.value_counts().plot(kind="bar", figsize=(10,6))
ax.set_xlabel("Registration Year")
ax.set_ylabel("User")
ax.set_title('Graph showing Registration Year of Used_Car')
```

```python
pt.pie(car['Registration_Year'].value_counts(), labels =car['Registration_Year'].unique(), autopct="%0.1f%%", radius=3)
pt.title("Graph showing Registration Year of Used_Car")
pt.show()
```

```python
ax = car['Fuel type'].value_counts().plot(kind="bar", figsize=(10,6))
ax.set_xlabel("Fuel type")
ax.set_ylabel("Number of User")
ax.set_title("Graph showing Fuel Type of Used_Car")
```

```python
explode = (0.01, 0.01, 1, 1, 1, 2)
pt.pie(car['Fuel type'].value_counts(),explode=explode, labels =car['Fuel type'].unique(), autopct="%0.1f%%")
pt.title("Graph showing Fuel Type of Used_Car")
pt.show()
```

```python
ax = car['Body type'].value_counts().plot(kind="bar", figsize=(10,6))
ax.set_xlabel("Body type")
ax.set_ylabel("Number of User")
ax.set_title("Graph Showing Body Type of Used_Car")
```

```python
explode = (0.01, 0.01, 0.01, 0.01, 0.01, 0.02, 0.02, 0.5, 2, 4)
pt.pie(car['Body type'].value_counts(), labels =car['Body type'].unique(), explode=explode, autopct="%0.1f%%", radius=1.1)
pt.title("Graph Showing Body Type of Used_Car")
pt.show()
```
